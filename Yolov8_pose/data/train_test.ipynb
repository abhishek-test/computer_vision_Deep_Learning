```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

def run_pose_inference_on_image_notebook(model_path, image_path):
    """
    Perform pose inference on a single image using a custom YOLOv8 pose model and display results in a Jupyter Notebook.
    
    Parameters:
        model_path (str): Path to the custom YOLOv8 model (.pt file).
        image_path (str): Path to the input image file.
    """
    # Load the custom YOLOv8 model
    model = YOLO(model_path)
    
    # Read the input image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to read image from {image_path}.")
        return
    
    # Convert BGR to RGB for Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(image)

    # Visualize the results
    annotated_image = results[0].plot()

    # Convert BGR to RGB for Matplotlib display
    annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

    # Display the original and annotated images side by side
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image_rgb)
    plt.axis("off")
    
    plt.subplot(1, 2, 2)
    plt.title("Pose Inference")
    plt.imshow(annotated_image_rgb)
    plt.axis("off")
    
    plt.show()


# Example usage in a Jupyter Notebook
# Replace 'custom_model.pt' with the path to your YOLOv8 model and 'input_image.jpg' with the image path.
run_pose_inference_on_image_notebook(
    model_path="C:\\Abhishek Data\\VSCode_Workspace\\Python\\DL_Practice\\Yolov8_pose\\data\\runs\\pose\\train2\\weights\\best.pt",
    image_path="2024-10-11_15-41-50-411363_door_volume.jpg"  # Replace with your image file path
)
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

def run_pose_inference_on_image_notebook(model_path, image_path):
    """
    Perform pose inference on a single image using a custom YOLOv8 pose model and display results in a Jupyter Notebook.
    
    Parameters:
        model_path (str): Path to the custom YOLOv8 model (.pt file).
        image_path (str): Path to the input image file.
    """
    # Load the custom YOLOv8 model
    model = YOLO(model_path)
    
    # Read the input image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to read image from {image_path}.")
        return
    
    # Get original dimensions
    original_height, original_width = image.shape[:2]

    # Convert BGR to RGB for Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(image)

    # Visualize the results
    annotated_image = results[0].plot()

    # Resize the annotated image to match the original dimensions
    annotated_image_resized = cv2.resize(annotated_image, (original_width, original_height), interpolation=cv2.INTER_LINEAR)

    # Convert BGR to RGB for Matplotlib display
    annotated_image_rgb = cv2.cvtColor(annotated_image_resized, cv2.COLOR_BGR2RGB)

    # Display the original and annotated images side by side
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image_rgb)
    plt.axis("off")
    
    plt.subplot(1, 2, 2)
    plt.title("Pose Inference (Resized)")
    plt.imshow(annotated_image_rgb)
    plt.axis("off")
    
    plt.show()

# Example usage in a Jupyter Notebook
# Replace 'custom_model.pt' with the path to your YOLOv8 model and 'input_image.jpg' with the image path.
run_pose_inference_on_image_notebook(
    model_path='C:\\Abhishek Data\\VSCode_Workspace\\Python\\DL_Practice\\Yolov8_pose\\data\\runs\\pose\\train2\\weights\\best.pt',
    image_path='2024-10-11_15-41-50-411363_door_volume.jpg'  # Replace with your image file path
)
```
